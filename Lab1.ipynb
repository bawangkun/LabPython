```python
#Q30: Input and String Operations

# ask user input
word = input("Enter a word: ")

# convert to upper,lower,reverse
upperWord = word.upper()
lowerWord = word.lower()
reverseWord = word[::-1] #[start:stop:step]

print("Uppercase:", upperWord)
print("Lowercase:", lowerWord)
print("Reversed:", reverseWord)
```

```python
# Q31: Basic Arithmetic Operations
# input two numbers seperate by space
num1, num2 = input("Enter two numbers seperated by space: ").split()

# convert to int
num1 = int(num1)
num2 = int(num2)

#perform basic arithmetic
sum = num1 + num2
diff = num1 - num2
product = num1 * num2
quotient = num1 / num2

#print
print("Sum:", sum)
print("Difference:", diff)
print("Product:", product)
print("Quotient:", quotient)
```

```python
#Q32: Exponentiation

#ask user input
num = input("Enter a number:")

# convert to int
num = int(num)

#calculate square and cube
square = num ** 2
cube = num ** 3

#print
print(num, "squared is", square)
print(num, "cubed is", cube)
```

```python
#Q33: Floor Division and Modulus

# ask user input
num1, num2 = input("Enter two numbers seperated by space: ").split()

#convert to int
num1 = int(num1)
num2 = int(num2)

# calculate floor division and modulus
floor_division = num1 // num2
modulus = num1 % num2

# print
print("Floor Division:", floor_division)
print("Modulus:", modulus)

```

```python
#Q37: Calculating Percentage

# ask input
marks_obtained = int(input("Enter marks obtained: "))
total_marks = int(input("Enter total marks: "))

# calculate percentage
percentage = (marks_obtained / total_marks) * 100

# print
print("Percentage:", percentage)
```

```python
#Q38: Simple Interest Calculation

# ask input
principal = int(input("Enter principal: "))
rate = int(input("Enter rate: "))
time = int(input("Enter time: "))

# calculate simple interest
simple_interest = (principal * rate * time) / 100

# print
print(simple_interest)
```
