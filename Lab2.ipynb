```python
#Q7: Find the GCD of two numbers using a while loop
#GCD = largest int num can divide for 2 or more numbers w/o remainder

num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

smaller = 0
if(num1<num2):
  smaller = num1
else:
  smaller = num2

gcd = 0
i = 1
while i<smaller+1:
  if (num1%i == 0) and (num2%i == 0):
    gcd = i
  i += 1

print("GCD is:", gcd)
```

```python
#Q2: Write a Python function that takes a number as input
#and returns True if it is prime, otherwise False.

print("Prime number checker.")

number = int(input("Enter a number: "))

def is_prime(n):
  if n < 2:
    return False

  for i in range(2,n):
    if n%i == 0:
     return False

  return True

print("prime number:",is_prime(number))
```

```python
#Q9: Write a Python function that returns
#the maximum of three given numbers.

def getNumber(n1,n2,n3):
  if n1 > n2 and n1 > n3:
    maximum = n1
  elif n2 > n1 and n2 > n3:
    maximum = n2
  else:
    maximum = n3
  return maximum


num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
num3 = int(input("Enter third number: "))

print("The maximum number:", getNumber(num1,num2,num3))
```
